```python
import plotly_express as px
```

```python
gapminder = px.data.gapminder()
gapminder
```

```python
gapminder2007 = gapminder.query('year == 2007')
gapminder2007
```

```python
fig = px.scatter(gapminder2007, x = "gdpPercap", y = "lifeExp")
fig.show()
```

ES INTERACTIVO

```python
fig = px.scatter(gapminder2007, x = "gdpPercap", y = "lifeExp", color = "continent")
fig.show()
```

```python
fig = px.scatter(gapminder2007, x = "gdpPercap", y = "lifeExp", color = "continent", size = "pop", size_max = 60, hover_name = "country")
fig.show()
```

```python
fig = px.scatter(gapminder2007, x = "gdpPercap", y = "lifeExp", color = "continent", size = "pop", size_max = 60, hover_name = "country", labels =  dict(gdpPercap = "GDP per capita"))
fig.show()
```

```python
fig = px.scatter(gapminder2007, x = "gdpPercap", y = "lifeExp", color = "continent", size = "pop", size_max = 60, hover_name = "country", facet_col = "continent", log_x = True)
fig.show()
```
